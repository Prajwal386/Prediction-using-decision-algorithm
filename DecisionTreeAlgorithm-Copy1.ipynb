```python
#Importing all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the data file
iris_dataset=pd.read_csv('Iris.csv')
```

```python
#Describing the head of dataset
iris_dataset.head()
```

```python
#Describing the dataset
iris_dataset.describe()
```

```python
iris_dataset.info()
```

# Visualisation

```python
#Plotting distribution of data
plt.figure(figsize=(10,5))
sns.scatterplot(data=iris_dataset, s=100, alpha=0.7)
plt.show()
```

# Corelation

```python
iris_dataset.corr()
sns.heatmap(iris_dataset.corr(),vmax=1,vmin=-1,annot=True,cmap='crest')
plt.show()
```

```python
#Checking for NULL values
iris_dataset.isnull().sum()
```

We can infer that there are no null values present in the dataset

```python
#Extracting independent & dependent variables from dataset
y=iris_dataset.iloc[:,-1]
x=iris_dataset.iloc[:,:-1]
```

```python
#Splitting the dataset into training & test split 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
#Fitting the model into training data set
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(x_train,y_train)
```

```python
#Predicting the test set 
y_predict=classifier.predict(x_test)
```

```python
y_predict
```

```python
#Tabulating the actual & predicted values
pd.DataFrame({'Actual':y_test,'predicted':y_predict})
```

```python
#Evaluating  Performance metrics
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

# Visualisation of decision tree

## Text representation

```python
from sklearn import tree
text_representation = tree.export_text(classifier)
print(text_representation)
```

#### Plot tree with tree plot

```python
features=iris_dataset.columns.to_list()[:-1]
target=iris_dataset.Species.unique()
print(f'features: {features}\n target: {target}')
```

```python
features=iris_dataset.columns.tolist()[:-1]
target=iris_dataset['Species'].unique()
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(classifier, 
                   feature_names=features,  
                   class_names=target,
                   filled=True)
```
